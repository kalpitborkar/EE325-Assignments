```python
import random
import numpy as np
```

```python
def choice(px, py, pz):
    r = random.random()
    
    if r<= px:
        return 1
    elif r<= px+py:
        return 2
    else:
        return 3
```

```python
def recur(pa, pb, pc, N, val, prob, temp):
    global lst1
    global lst2
    global lst3
    if temp == N:
        lst1.append(prob * val)
        lst2.append(val)
        lst3.append(prob*val**2)
        
        
    else:
        recur(pa, pb, pc, N, val+1, prob*pa, temp+1)



        recur(pa, pb, pc, N, max(0, val-1), prob*pb, temp+1)



        recur(pa, pb, pc, N, val, prob*pc, temp+1)
```

```python
print ('\033[1m' + 'a'+'\033[0m')
print()
a= .7
b = .9
q = [2]
p = 100
p_list = list(range(1, 16))
n1 = []
n2 = []
n3 = []
E = [2]
t_avg, ensemble_avg = 0, 0
p1, p2, p3 = a*(1-b), b*(1-a), a*b + (1-a)*(1-b)
for i in range(p):
    k = choice(p1, p2, p3)

    if k == 1:
        q.append(q[-1]+1)
    elif k==2:
        q.append(max(0, q[-1]-1))
    elif k==3:
        q.append(q[-1])


    A = [x for x in q if x <=100000]
    B = [y for y in A if y%p == i]
    #print(len(B)/len(A), end = "   ")
    
for p in p_list:
    lst1 = []
    lst2 = []
    lst3 = []
    recur(p1, p2, p3, p, 3, 1, 0)
    n1.append(np.sum(np.array(lst1)))
    n2.append(np.average(np.array(lst2)))
    n3.append(np.sum(np.array(lst3)))
print("Expectation=" ,"\n", n1, "\n\n", "Ensemble_avg =", "\n", n2, "\n\n") 
print("Q_list:", "\n",  q)
print()
t_avg = np.average(np.array(q))
print("Time avg =", t_avg)
```
