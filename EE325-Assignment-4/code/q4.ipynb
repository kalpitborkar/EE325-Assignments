```python
from array import *
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
from math import comb
```

```python
value = [0,1,2,3,4,5,6] #set of values
```

```python
def simulation(p,value):
    x = [0] #sequence Xn
    x.append(random.choice(value)) #first value is picked randomly
    
    #probability variables
    prob_x = [0]
    curr_prob = 1
    
    for n in range(1000): #length of n = 100 (x1 x2 ... x100) (x1 is already chosen)
        i = x[-1]
        p_curr = p[i]
        next_element = random.choices(value, weights = p_curr, k=1)
        x.append(next_element[0])
        
        #current probability
        for i in value:
            if(next_element[0]==i):
                #curr_prob = p_curr[i]*curr_prob
                curr_prob = p_curr[i]
            
            
        prob_x.append(curr_prob)
        
    #print(x)
    return [x,prob_x]
```
