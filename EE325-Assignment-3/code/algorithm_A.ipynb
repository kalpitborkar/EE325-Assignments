```python
from array import *
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
from math import comb
import math
import pandas as pd
```

```python


#N1 is a fixed constant and we need to iterate it from 1 to N-1
#N1 = 3
#n = N1/3
#n = 1

```

```python
#fix N1 < N
#toss each coin N1/3 times
```

```python
pa = [0.2, .45]
pb = [0.4, .5]
pc = [0.7, .58]
N = [20, 100, 1000, 5000]
best_N1 = [[], []]
correct_coin = [[], []]

for i in range(0, 2):
    for N in [20, 100, 100, 5000]:
        R_final= np.array([])
        correct_coin_temp = []
        #n = number of times each coin is being tossed
        for n in range(1, math.floor(N/3) + 1):
            R_for_n1 = np.array([])
            correct_coin_n1 = []
            
            for y in range(1000):

                temp = 0
                na = np.random.binomial(n, pa[i])

                nb = np.random.binomial(n, pb[i])

                nc = np.random.binomial(n, pc[i])


                max_heads = max([na,nb,nc])
                
                if max_heads == na:
                    best_p = pa[i]
                elif max_heads == nb:
                    best_p = pb[i]
                elif max_heads == nc:
                    best_p = pc[i]
                    temp = N - 3*n
                    

                #simulating N-N1 tosses
                nr = np.random.binomial(N-(3*n), best_p)

                R = na + nb + nc + nr
                R_for_n1 =np.append(R_for_n1,R) #contains 1000 values for a fixed n1
                correct_coin_n1.append(temp)

            temp_avg = np.average(R_for_n1)
            correct_coin_temp.append(np.average(np.array(correct_coin_n1)))
            
            R_final =np.append(R_final,temp_avg)


            #R_arr =np.append(R_arr,R)
            #best_N1 = (1+ np.argmax(R_arr))*3
        correct_coin[i].append(correct_coin_temp[np.argmax(R_final)])
        best_N1[i].append((1+ np.argmax(R_final))*3)

print(best_N1)
print(correct_coin)
```

```python
class_1= {"Case": 4* ["pa = .2, pb = .4, pc = .7"], "N": [20, 100, 1000, 5000], "Best_N1": best_N1[0],
          "Correct_no_of_coin":correct_coin[0]}
table_1 = pd.DataFrame(class_1)

class_2= {"Case": 4* ["pa = .45, pb = .5, pc = .58"], "N": [20, 100, 1000, 5000], "Best_N1": best_N1[1],
          "Correct_no_of_coin":correct_coin[1]}
table_2 = pd.DataFrame(class_2)

table_3 = pd.concat([table_1, table_2], ignore_index = True)
table_3.groupby(["Case", "N"]).first()
```
